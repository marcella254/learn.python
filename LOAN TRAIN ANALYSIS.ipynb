```python
data=pd.read_csv("loan-train.csv")


#checking for missing values
np.sum(data.isna())

#Gender-----------------------------------
#checking the comparison btwn gender and married
data[['Gender','Married']].value_counts()

#checking the rows with na"s
data[data['Gender'].isna()]

#most married gender is male and many females are not
data['Gender']=data['Gender'].fillna('Male')

#transform the data types
data['Gender']=data['Gender'].astype('category')

#Married--------------------
#check the total categories in married column
data['Married'].value_counts()
#check for nas
data[data['Married'].isna()]
data['Married']=data['Married'].fillna('Yes')
#transform datatype
data['Married']=data['Married'].astype('category')

#Dependents----------------------------------
#check relationship btn dependents and marriage
data[['Dependents','Married']].value_counts()
#check for nas\
data[data['Dependents'].isna()]
data['Dependents']=data['Dependents'].fillna(0)
#tranforming the variable
data['Dependents']=data['Dependents'].replace({'3+':3,"3":3})
data['Dependents']=data['Dependents'].astype('category')

#Self Employment--------------------------------
#checking nas
data[data['Self_Employed'].isna()]
data['Self_Employed']=data['Self_Employed'].fillna('No')

#transform the datatype
data['Self_Employed']=data['Self_Employed'].astype('category')

#loan Amount------------------------------
data['LoanAmount'].value_counts()
#checking nas
data[data['LoanAmount'].isna()]
#checking for distribution
#data.hist(column='LoanAmount')#IT IS SKEWED REPLACE nas with median
data.describe()
data['LoanAmount']=data['LoanAmount'].fillna(128)

#Loan_Amount_Term----------------------------------
#checking  nas
data[data['Loan_Amount_Term'].isna()]
#checking for distribution
#data.hist(column='Loan_Amount_Term')#skewed replace with median
data['Loan_Amount_Term']=data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].median())

#credit history----------------------------------
#check nas
data[data['Credit_History'].isna()]

#transform
data['Credit_History']=data['Credit_History'].fillna(1)
data['Credit_History']=data['Credit_History'].astype('category')
data[['Education','Loan_Status','Property_Area']]=data[['Education','Loan_Status','Property_Area']].astype('category')
data.info()

#outliers for numerical data--------------------------
#data.boxplot(column='ApplicantIncome')#outliers
#data.hist(column='ApplicantIncome')#outliers:skewed
#normalisation
Normalised_ApplicantIncome=np.log(data['ApplicantIncome'])
data['Normalised_ApplicantIncome']=Normalised_ApplicantIncome

# data.boxplot(column='CoapplicantIncome')#:outliers
# data.hist(column='CoapplicantIncome')#:skewed
normalised_CoapplicantIncome=np.log(data['CoapplicantIncome'])
data['Normalised_CoapplicantIncome']=normalised_CoapplicantIncome
data['Normalised_CoapplicantIncome']=data['Normalised_CoapplicantIncome'].replace({float('-inf'):0})

#data.boxplot(column='LoanAmount')#ouliers
Normalised_LoanAmount=np.log(data['LoanAmount'])
data['Normalised_LoanAmount']=Normalised_LoanAmount

#data.boxplot(column='Loan_Amount_Term')#outliers
Normalised_Loan_Amount_Term=np.log(data['Loan_Amount_Term'])
data['Normalised_Loan_Amount_Term']=Normalised_Loan_Amount_Term

data.to_csv('C:\\Users\\marcella\\Downloads\\loan_train.csv',index=False)
```
